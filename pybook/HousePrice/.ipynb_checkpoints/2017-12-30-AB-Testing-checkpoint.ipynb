Remember that the main purpose of control variables is to make sure that you have a good match between
treatment and control units. So that the comparison being made
in the experiment is a good one.

Five steps should be followed to determine which control variables should be used to set up the experiment.

First, come up with **a list of potential variables.**

Next, determine whether you have data for those variables **(Is variable data available?)**.

Then make sure that the **connection between the target variable and each control variable is logical**.

Next, test the **correlation between the control and target variables.**

And lastly, **test correlation between other control variables** to avoid similar variables.


So how many control variables should you use?

Well, there's no hard rule.

It could be as few as one, or as many as five and sometimes more.

Having too many control variables is usually worse for an experiment than having too few.

If the experiment was done as a randomized design the advantage would be that the experiment would be considerably easier and cheaper to set up and analyze. The disadvantage would be that it would not directly account for possible bias.

If the experiment was done as a matched pair design it would directly control for identified sources of bias (not just leaving random chance to deal with them). The drawback is that this type of experiment is more costly and time intensive to set up and analyze.
